## 1. NumPy 基礎 NumPy Basics
NumPy (Numerical Python) 是 Python 數值計算的基石。

```python
# NumPy 基礎操作
import numpy as np

# 建立陣列 Create arrays
a = np.array([1, 2, 3, 4, 5])
b = np.linspace(0, 1, 5)  # 等距數列

print("陣列 a:", a)
print("等距數列 b:", b)
print("形狀 Shape:", a.shape)
print("平均值 Mean:", a.mean())
print("標準差 Std:", a.std())

# 矩陣運算 Matrix operations
M = np.array([[1, 2], [3, 4]])
print("\n矩陣 M:\n", M)
print("轉置 Transpose:\n", M.T)
print("行列式 Determinant:", np.linalg.det(M))
```

## 2. Pandas 基礎 Pandas Basics
Pandas 用於結構化資料的讀取、清洗與分析。

```python
# Pandas 基礎操作
import pandas as pd

# 建立 DataFrame
data = {
    '姓名': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    '年齡': [23, 25, 22, 28, 24],
    '成績': [85, 92, 78, 95, 88],
    '科系': ['資工', '資管', '資工', '電機', '資管']
}

df = pd.DataFrame(data)
print("DataFrame 預覽：")
print(df)
print("\n基本統計 Descriptive Statistics：")
print(df.describe())
print("\n各科系平均成績 Mean Score by Department：")
print(df.groupby('科系')['成績'].mean())
```

## 3. Matplotlib 基礎視覺化 Basic Visualization

```python
# Matplotlib 基礎繪圖
import matplotlib.pyplot as plt
import numpy as np

# 設定中文字型（如果需要）
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei', 'SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 1. 折線圖 Line Plot
x = np.linspace(0, 2 * np.pi, 100)
axes[0].plot(x, np.sin(x), label='sin(x)', color='#2563eb')
axes[0].plot(x, np.cos(x), label='cos(x)', color='#dc2626')
axes[0].set_title('折線圖 Line Plot')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# 2. 散佈圖 Scatter Plot
np.random.seed(42)
x_scatter = np.random.randn(100)
y_scatter = 2 * x_scatter + np.random.randn(100) * 0.5
axes[1].scatter(x_scatter, y_scatter, alpha=0.6, c='#7c3aed')
axes[1].set_title('散佈圖 Scatter Plot')
axes[1].grid(True, alpha=0.3)

# 3. 長條圖 Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
axes[2].bar(categories, values, color=['#2563eb', '#7c3aed', '#dc2626', '#059669'])
axes[2].set_title('長條圖 Bar Chart')

plt.tight_layout()
plt.show()
```

## 4. 小試身手：ML 資料集初探
讓我們用 scikit-learn 內建的鳶尾花資料集 (Iris Dataset) 做一個快速的資料探索。

```python
# 載入鳶尾花資料集 Load Iris Dataset
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt

iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

print("資料集大小 Dataset Shape:", df_iris.shape)
print("\n前 5 筆資料 First 5 rows:")
print(df_iris.head())
print("\n各類別數量 Class Distribution:")
print(df_iris['species'].value_counts())

# 簡單散佈圖
fig, ax = plt.subplots(figsize=(8, 6))
colors = {'setosa': '#2563eb', 'versicolor': '#dc2626', 'virginica': '#059669'}
for species, group in df_iris.groupby('species'):
    ax.scatter(group['sepal length (cm)'], group['sepal width (cm)'],
               label=species, alpha=0.7, color=colors[species])
ax.set_xlabel('花萼長度 Sepal Length (cm)')
ax.set_ylabel('花萼寬度 Sepal Width (cm)')
ax.set_title('鳶尾花資料集 Iris Dataset')
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()
```
